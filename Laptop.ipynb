```python
df['graphic_card_gb'] = df.graphic_card_gb.str.replace(' GB', '')
df['graphic_card_gb'] = df.graphic_card_gb.astype(int)
```

```python
df.warranty.value_counts()
```

```python
def warranty(num):
    if num == '1 year': return 1
    elif num == '2 years': return 2
    elif num == '3 years': return 3
    else: return 0
```

```python
df.warranty = df.warranty.apply(warranty)
```

```python
df.head()
```

```python
df['rating'].value_counts()
```

```python
def rating(star):
    if star == '1 star': return 1
    elif star == '2 stars': return 2
    elif star == '3 stars': return 3
    elif star == '4 stars': return 4
    else: return 5
```

```python
df.rating = df['rating'].apply(rating)
```

```python
df.drop(columns = ['os_bit','processor_gnrtn','rating', 'Number of Reviews', 'Number of Ratings'], axis = 1, inplace = True)
```

```python
df
```

```python
X = df.drop('Price', axis=1)
```

```python
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.26, random_state = 4)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trans = ColumnTransformer(transformers = [('ohe', OneHotEncoder(sparse_output = False, drop = 'first'), [0, 1, 2, 4, 7, 9, 11, 12])], remainder = 'passthrough')
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([('transformation', trans)])
```

```python
pipe
```

```python
X_train_transformed = pipe.fit_transform(X_train)
X_test_transformed = pipe.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train_transformed, columns=pipe.named_steps['transformation'].get_feature_names_out())
X_test = pd.DataFrame(X_test_transformed, columns=pipe.named_steps['transformation'].get_feature_names_out())
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

model = RandomForestRegressor()
model.fit(X_train, y_train)
model_pred = model.predict(X_test)
```

```python
r2_score(y_test, model_pred)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(estimator = model, X = X_train, y = y_train, scoring = 'r2', cv = 4).mean()
```

```python
from sklearn.model_selection import GridSearchCV

parameters = [{'max_samples':[0.7, 0.77, 0.85, 0.92, 1], 'max_depth':[9, 11, 13, 15, 17]}]
search = GridSearchCV(estimator = model, param_grid = parameters, scoring = 'r2', cv = 4)
search = search.fit(X_train, y_train)
```

```python
search.best_score_
```

```python
search.best_params_
```

```python
model.fit(X_train, y_train)
```

```python
model_pred = model.predict(X_test)
```

```python
r2_score(y_test, model_pred)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

boost = GradientBoostingRegressor(n_estimators = 250)
boost.fit(X_train, y_train)
boost_pred = boost.predict(X_test)
```

```python
r2_score(y_test, boost_pred)
```
